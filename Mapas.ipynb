```python
mapa=folium.Map(location=[9.9707974,-84.1292153],zoom_start=16,titles='stamenterrain')
C_Benjamin='<b>Campus Benjamín Núñez</b>'
C_Omar_Dengo='<b>Campus Omar Dengo</b>'
C_Coto='<b>Campus Coto</b>'
C_Liberia='<b>Campus Liberia</b>'
C_Perez_Zeledon='<b>Campus Perez Zeledon</b>'

folium.Marker(location=[9.9707974,-84.1292153],popup=C_Benjamin,icon=folium.Icon(color="purple",icon="ok-sign")).add_to(mapa)
folium.Circle(location=[9.9707974,-84.1292153],popup=C_Benjamin,color='black', weight=6,fill_color='red',fill_opacity=0.5,tooltip='Benjamín Núñez').add_to(mapa)

folium.Marker(location=[9.9990878,-84.1117806],popup=C_Omar_Dengo,icon=folium.Icon(color="red",icon="ok-sign")).add_to(mapa)
folium.Circle(location=[9.9990878,-84.1117806],popup=C_Omar_Dengo,color='black', weight=6,fill_color='blue',fill_opacity=0.5,tooltip='Omar Dengo').add_to(mapa)

folium.Marker(location=[8.5733876,-82.8727255],popup=C_Coto,icon=folium.Icon(color="green",icon="ok-sign")).add_to(mapa)
folium.Circle(location=[8.5733876,-82.8727255],popup=C_Coto,color='black', weight=6,fill_color='red',fill_opacity=0.5,tooltip='Coto').add_to(mapa)

folium.Marker(location=[10.6174646,-85.4516771],popup=C_Liberia,icon=folium.Icon(color="orange",icon="ok-sign")).add_to(mapa)
folium.Circle(location=[10.6174646,-85.4516771],popup=C_Liberia,color='black', weight=6,fill_color='red',fill_opacity=0.5,tooltip='Liberia').add_to(mapa)

folium.Marker(location=[9.379744,-83.6916421],popup=C_Perez_Zeledon,icon=folium.Icon(color="blue",icon="ok-sign")).add_to(mapa)
folium.Circle(location=[9.379744,-83.6916421],popup=C_Perez_Zeledon,color='black', weight=6,fill_color='red',fill_opacity=0.5,tooltip='Perez Zeledon').add_to(mapa)

minimap=MiniMap(title_layer='stamenterrain',position='botomright')
mapa.add_child(minimap)
```
